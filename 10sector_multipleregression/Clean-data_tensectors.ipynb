```python
import pandas as pd
from functools import reduce
```

```python
#read in data from csv's
file_path = "data\AllSectors_2019.csv"

allsect_df = pd.read_csv(file_path)

allsect_df
```

```python
#grab unique sectors
allsect_df['Sector'].unique()
```

```python
#1 return only data from Consumer Staples
consumer_df = allsect_df.loc[allsect_df['Sector'] == 'Consumer Staples']

#grab only relevant data
consumer_df = consumer_df.loc[:,['Date','Close']]
consumer_df1 = consumer_df.rename(columns={"Close": "Consumer Staples"})
n_con_df = consumer_df1.round({'Consumer Staples': 2})

n_con_df.head()
```

```python
#2 return only data from Information Technology
info_df = allsect_df.loc[allsect_df['Sector'] == 'Information Technology']

#grab only relevant data
info_df = info_df.loc[:,['Date','Close']]
info_df1 = info_df.rename(columns={"Close": "Information Technology"})
n_info_df = info_df1.round({'Information Technology': 2})

n_info_df.head()
```

```python
data_frames = [n_con_df,n_info_df,n_amer_df,n_con1_df,n_uti_df,n_real_df,\
                    n_health_df,n_materials_df,n_energy_df,n_tele_df,n_ind_df,n_sp500_df]

df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['Date'],
                                            how='outer'), data_frames)

df_merged
```
